# Import Dataset

```python
import pandas as pd
dataset=pd.read_csv('D:/machine learning/m7/water_potability.csv')
dataset
```

# Mengisi Missing Value dengan rata-rata

```python
data=dataset.fillna(dataset.groupby('Potability').transform('mean'))
data
```

# Menghapus baris yang berisi Missing Value

```python
data=dataset.dropna()
data
```

# Normalization Range 0 - 1

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler(feature_range=(0,1))
data=sc.fit_transform(data)
data
```

Merapikan bentuuk data kembali seperti semula 

```python
data=pd.DataFrame(data,columns=['ph','Hardness','Solids','Chloramines','Sulfate','Conductivity','Organic_carbon','Trihalomethanes','Turbidity','Potability'])
data
```

# Sort Data

```python
#ascending is default
ak=data.sort_values(by=['ph'])
ak
```

Sort descending

```python
#descending
dk=data.sort_values(by='ph',ascending=False)
dk
```

# Menyimpan dataset 

```python
dk.to_csv('water_potability_desc.csv')
```
